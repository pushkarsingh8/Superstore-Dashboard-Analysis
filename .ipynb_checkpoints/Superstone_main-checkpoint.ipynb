```python
import pandas as pd
import matplotlib.pyplot as mt
df = pd.read_csv('C:/Users/1push/Downloads/dataset/Superstone.csv',encoding = 'latin-1')
df['year'] = pd.to_datetime(df['Order Date']).dt.year
```

```python
fig = mt.figure(figsize = (12,8),constrained_layout = True,facecolor = 'lightgray')
fig.suptitle('US Superstore Dashboard (2014–2018)', fontsize=16, fontweight='bold')


gs = mt.GridSpec(2,3,figure = fig)




#first chart display about Yearly sales & profit 

ax1 = fig.add_subplot(gs[0,0:2])

df['year'] = pd.to_datetime(df['Ship Date']).dt.year
year_salesprofit = df.groupby('year')[['Sales','Profit']].sum().reset_index() #yearly sales / profit creating line chart


ax1.plot(year_salesprofit['year'],year_salesprofit['Sales'],label = 'Sales by Year',marker = 'o')
ax1.plot(year_salesprofit['year'],year_salesprofit['Profit'],label = 'Profit by Year',marker = '*')

mt.gca().set_facecolor('lightblue')

ax1.set_title('Profit & Sales By Year')
ax1.set_xlabel('Year')
ax1.set_ylabel('Profit & Sales')

ax1.legend(loc = 'best')
ax1.set_xticks(year_salesprofit.year)


for i in range(len(year_salesprofit)):
     ax1.text(year_salesprofit['year'][i],year_salesprofit['Profit'][i],s = str(year_salesprofit['Profit'][i]),ha = 'left',va = 'top',color = 'green')
     ax1.text(year_salesprofit['year'][i],year_salesprofit['Sales'][i],s = str(year_salesprofit['Sales'][i]),ha = 'center',va = 'bottom',color = 'blue')
   
mt.grid()



#Second chart show 'Cities with Highest Number of Customers'"
ax2 = fig.add_subplot(gs[1,0:2])

city_custmr = df['City'].value_counts().reset_index().head(10) #top 10 customer come city in u.s.
ax2.set_title('Top 10 Cities by Customer Count')

top_cities = city_custmr.sort_values(by='count', ascending=False).head(10)
max_count = city_custmr['count'].max()

colors = ['green' if count == max_count else '#B8860B' for count in top_cities['count']] #set target on top sales

ax2.barh(city_custmr['City'],city_custmr['count'], label = 'Cities High Sales',color = colors)

for i in range(len(city_custmr)):
     ax2.text(city_custmr['count'][i],city_custmr['City'][i],s = str(city_custmr['count'][i]),ha = 'right',va = 'center',color = 'white')
     
mt.gca().set_facecolor('lightblue')
ax2.legend(loc = 'best')
ax2.set_xlabel('Customers Count')
ax2.set_ylabel('City')



#Third chart show Regionaly wise customer repeats
ax3 = fig.add_subplot(gs[:,2])

region_rcr = df['Region'].value_counts().reset_index() 
region_rcr['Percent'] = (region_rcr['count']/region_rcr['count'].sum())*100 #rcr it's 3 where we see pie chart

explode = [0,0,0,0.1]
ax3.pie(region_rcr.Percent,labels = region_rcr.Region,autopct = '%0.1f%%',explode = explode)
ax3.set_title('Regional Share of Total Customers')

fig.savefig('superstone_dashboard.png', dpi=300, bbox_inches='tight')

mt.show()
```
